## Cleaning of the dataset

```python
#creating a backup copy in case the data gets corrupted while cleaning
```

```python
backup=data.copy()
```

```python
data=data[data['year'].str.isnumeric()] #removing the rows whose year data is not numeric
```

```python
data.loc[:,'year']=data['year'].astype(int)  #now we have data in numeric from only so convert datatype into int 
```

```python
data['year'].info()
```

```python
# now year's dtype is int  and have 842 rows in total
```

```python
data=data[data['Price']!='Ask For Price']
```

rows are removed which had ask for proce instead of price in *Price* column

```python
data['Price'] = data['Price'].astype(str).str.replace(',', '').astype(int)
```

```python
#Price dtype is converted to int and ',' is replaced to empty string as we can;t have ',' in int
```

```python
data['kms_driven']=data['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
data=data[data['kms_driven'].str.isnumeric()]
```

```python
data['kms_driven']=data['kms_driven'].astype(int)
```

```python
data=data[~data['fuel_type'].isna()] #removing the values in fuel_type which are nan
```

```python
## shorting the names(keep 1st letter of each name)
```

```python
data['name']=data['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
data
```

```python
# we need to reset index of rows as we have deleted some of the rows
```

```python
data.reset_index()
```

```python
data.describe()
```

```python
#finding cars which are more than 60lakhs
data=data[data['Price']<6e6].reset_index(drop=True)
```

## Model

```python
#x=all columns values except price
#y=price values
X=data.drop(columns='Price')
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe.categories_
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough',force_int_remainder_cols=False )
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
#finding random state at which it gives maximum r2 score 
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
#max r2_score is 0.845 at random state 433
```

```python
#putting this random state in train test split code
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=np.argmax(scores),test_size=0.2)
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
